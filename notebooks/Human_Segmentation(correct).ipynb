# **Defining model**

---

После изучения общепринятых подходов к проблеме сегментации изображений было принято решение подробнее изучить архитектуру Mask R-CNN и в дальнейшем её применить к задаче.


---



**Mask R-CNN** представляет собой двухэтапную структуру: на первом этапе изображение сканируется и генерируются области, в которых может содержаться объект.На втором этапе эти области классифицируются и генерируются  ограничивающие рамки,в которых содержится объект, затем генерируются маски.


Эта архитектура была представлена в 2017 году ([ Mask R-CNN paper](https://arxiv.org/abs/1703.06870)).Mask R-CNN это продолжение архитектуры Faster R-CNN, предложенной этими же авторами,путем добавления метода сегментации изображений.

![alt text](https://miro.medium.com/max/1400/1*IWWOPIYLqqF9i_gXPmBk3g.png)




Mask R-CNN содержит следующие модули:

## **1. Backbone**

Это стандартная сверточная нейронная сеть (обычно ResNet50 или ResNet101), которая служит в качестве экстрактора признаков. Ранние слои "обнаруживают" объекты низкого уровня (края и углы), а более поздние слои  "обнаруживают" элементы более высокого уровня (автомобиль, человек, небо).
Проходя через backbone , изображение преобразуется из 1024x1024x3  в карту признаков(feature map) формы 32x32x2048. Этот feature map становится входом для следующих модулей.


**Feature Pyramid Network**

FPN улучшает стандартную пирамиду извлечения объектов, добавляя вторую пирамиду, которая берет признаки(features) высокого уровня от первой пирамиды и передает их вниз в нижние слои. Это позволяет  на каждом уровне иметь доступ к признакам как нижнего, так и более высокого уровня.

![FPN](https://im0-tub-ru.yandex.net/i?id=7cdb636f59d1123e6f3638193333d955&n=13&exp=1)

## **2. Region Proposal Network (RPN)**

RPN это нейросеть,которая в режиме скользящего окна сканирует feature map,полученный из backbone, в определенных выбранных областях(anchors)

На выходе из RPN получается:

**Anchor class:** один из двух классов: передний план или фон.Если класс anchor определен как "передний план", то это означает что скорее всего в этом anchor есть объект.

**Bounding Box Refinement:** RPN рассчитывает процент изменений по ширине и высоте, для того чтобы уточнить расположение ограничивающий рамки объекта


На основе этих предсказаний определяются регионы интересов (regions of interest - RoI)(выбираются anchor,которые содержат объекты и определяются их размеры)

![alt text](https://www.codeproject.com/KB/AI/1280553/d8efdd62-230a-421c-8647-ececed686276.Jpeg)

## **3. ROI Classifier & Bounding Box Regressor**

На этоп этапе ROI Classifier проходит по RoI регионам, полученным от RPN и генерирует предсказания для каждого:

1)Class: класс объекта в ROI. В отличие от RPN, который имеет два класса (FG / BG), здесь могут предсказываться различные классы  (человек, машина, стул и т. Д.).А также может генерировать фоновый класс(BG), который вызывает отбрасывание ROI.

2)Bounding Box Refinement: уточненное расположение и размер ограничительной рамки для выделения объекта


![alt text](https://miro.medium.com/max/933/1*xQYuM_9mu5kt8nNN8Ms2TQ.png)



**ROI Pooling**

ROI Pooling вырезает RoI на  feature map  и изменяет его размер до необходимого .
При этом применяется метод, который авторы назвали ROIAlign(вырезается область и изменяется её размер до необходимого путем применяется билинейной интерполяции)



## **4. Segmentation Masks**

На этом этапе берутся  положительные области, полученные от  ROI Classifier, и генерируются маски для них. 

![alt text](https://www.learnopencv.com/wp-content/uploads/2019/06/res_mrcnn_cars-300x183.jpg)

# **Defining the Dataset's and Dataloader's**

Чтобы воспользоваться уже  предобученной сетью Mask R-CNN,реализованной в PyTorch'е создадим тренировочный и валидационный датасеты, а также Dataloader's,которые позволяет итерироваться по датасету в процессе обучения модели.

Классы датасетов наследуются от стандартного **torch.utils.data.Dataset class**, но нужно переопределить методы **\_\_len\_\_** and **\_\_getitem\_\_**.


Чтобы датасет использовался при обучении Masr R-Cnn, его особенностью является то, что метод     **\_\_getitem\_\_** должен возвратить:


- **image**:  PIL Image размера (H, W)



- **target**: словарь, содержащий следующее:

 1. **boxes** (FloatTensor[N, 4]): координаты ограничивающих рамок для объектов 

 2. **labels** (Int64Tensor[N]): метка класса для каждой ограничивающий рамки

 3. **image_id** (Int64Tensor[1]): уникальный идентификатор изображения.

 4. **area** (Tensor[N]): Площадь ограничивающий рамки.

### Создадим датасеты и добавим функцию аугментации/трансформации данных.


```python
import transforms as T
import utils
def get_transform(train):
  """
   Function for data augmentation
   
  Parameters
    ----------
    train : bool
        True: using for train dataset
        False: using for valid dataset
        
    Returns
    -------
           Augmentation method to apply to the dataset element
           
           
  """
  
  transforms = []
  # convert PIL image  - > PyTorch Tensor
  transforms.append(T.ToTensor())
  if train:
      # apply  a random flip for the image and its mask
      #for data augmentation
      transforms.append(T.RandomHorizontalFlip(0.5))
  return T.Compose(transforms)
```

```python
class PersonDataset_train(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms
        # load all image files, sorting them to
        # ensure that they are aligned
        self.imgs = list(sorted(os.listdir(os.path.join(root, "train"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "train_mask"))))

    def __getitem__(self, idx):
        # load images ad masks
        img_path = os.path.join(self.root, "train", self.imgs[idx])
        mask_path = os.path.join(self.root, "train_mask", self.masks[idx])
        img = Image.open(img_path).convert("RGB")
       
      
        mask = Image.open(mask_path)

        mask = np.array(mask)
        # instances are encoded as different colors
        obj_ids = np.unique(mask)
        # first id is the background, so remove it
        obj_ids = obj_ids[1:]

        # split the color-encoded mask into a set
        # of binary masks
        masks = mask == obj_ids[:, None, None]

        # get bounding box coordinates for each mask
        num_objs = len(obj_ids)
        boxes = []
        for i in range(num_objs):
            pos = np.where(masks[i])
            xmin = np.min(pos[1])
            xmax = np.max(pos[1])
            ymin = np.min(pos[0])
            ymax = np.max(pos[0])
            boxes.append([xmin, ymin, xmax, ymax])

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        # there is only one class
        labels = torch.ones((num_objs,), dtype=torch.int64)
        masks = torch.as_tensor(masks, dtype=torch.uint8)

        image_id = torch.tensor([idx])
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0])
        # suppose all instances are not crowd
        iscrowd = torch.zeros((num_objs,), dtype=torch.int64)

        target = {}
        target["boxes"] = boxes
        target["labels"] = labels
        target["masks"] = masks
        target["image_id"] = image_id
        target["area"] = area
        target["iscrowd"] = iscrowd

        if self.transforms is not None:
            img, target = self.transforms(img, target)

        return img, target

    def __len__(self):
        return len(self.imgs)
```

```python
class PersonDataset_valid(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms
        # load all image files, sorting them to
        # ensure that they are aligned
        self.imgs = list(sorted(os.listdir(os.path.join(root, "valid"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "valid_mask"))))

    def __getitem__(self, idx):
        # load images ad masks
        img_path = os.path.join(self.root, "valid", self.imgs[idx])
        mask_path = os.path.join(self.root, "valid_mask", self.masks[idx])
        img = Image.open(img_path).convert("RGB")
        
        
        mask = Image.open(mask_path)

        mask = np.array(mask)
        # instances are encoded as different colors
        obj_ids = np.unique(mask)
        # first id is the background, so remove it
        obj_ids = obj_ids[1:]

        # split the color-encoded mask into a set
        # of binary masks
        masks = mask == obj_ids[:, None, None]

        # get bounding box coordinates for each mask
        num_objs = len(obj_ids)
        boxes = []
        for i in range(num_objs):
            pos = np.where(masks[i])
            xmin = np.min(pos[1])
            xmax = np.max(pos[1])
            ymin = np.min(pos[0])
            ymax = np.max(pos[0])
            boxes.append([xmin, ymin, xmax, ymax])

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        # there is only one class
        labels = torch.ones((num_objs,), dtype=torch.int64)
        masks = torch.as_tensor(masks, dtype=torch.uint8)

        image_id = torch.tensor([idx])
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0])
        # suppose all instances are not crowd
        iscrowd = torch.zeros((num_objs,), dtype=torch.int64)

        target = {}
        target["boxes"] = boxes
        target["labels"] = labels
        target["masks"] = masks
        target["image_id"] = image_id
        target["area"] = area
        target["iscrowd"] = iscrowd

        if self.transforms is not None:
            img, target = self.transforms(img, target)

        return img, target

    def __len__(self):
        return len(self.imgs)
```

# **Сreate and train model**

**Создадим модель**

```python
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from torchvision.models.detection.mask_rcnn import MaskRCNNPredictor

      
def get_instance_segmentation_model(num_classes):
    """
   Function for model creating
   
    Parameters
    ----------
    num_classes : int - number of classes
        
        
    Returns
    -------
       Pretrained model with  the last module(mask_predictor) 
       replaced by a new one
           
    """
    
  
    #загрузим Mask R-CNN модель, предтренированную на COCO датасете
    model = torchvision.models.detection.maskrcnn_resnet50_fpn(pretrained=True)

    
    #фичи для RoI классификатора
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    
    # часть преобученной модели заменим на  новую модель
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

    #получим число фичей для последнего модуля(mask_predictor)
    in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
    hidden_layer = 256
    # заменим  mask_predictor на новый
    model.roi_heads.mask_predictor = MaskRCNNPredictor(in_features_mask,
                                                       hidden_layer,
                                                       num_classes)

    return model
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# Определим количество классов(задний фон и человек)
num_classes = 2

# создадим модель
model = get_instance_segmentation_model(num_classes)
# положим модель на GPU
model.to(device)

# Зададим метод оптимизации
params = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.SGD(params, lr=0.005,
                            momentum=0.9, weight_decay=0.0005)


# добавим уменьшение скорости обучения в процессе обучения модели
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer,
                                               step_size=3,
                                               gamma=0.1)
```

```python
from engine import train_one_epoch, evaluate
# Запустим обучение
num_epochs = 10

for epoch in range(num_epochs):
    # для каждой эпохи выведем на печать значения метрик для 10 итераций
    train_one_epoch(model, optimizer, data_loader, device, epoch, print_freq=10)
    # обновим скорость обучения
    lr_scheduler.step()
    # оценим качество на валидационном датасете
    evaluate(model, data_loader_test, device=device)
```

```python
def dice_full(dataset):
    """
     Get average Dice score
     
    Parameters
    ----------
    dataset: (torch.utils.data.Dataset class) valid dataset
    
    Return:
    ----------
    average Dice score on valid dataset
    """

  model.eval()
  list_of_pred_mask = []
  list_of_true_mask = []
  for elem in range(dataset_test.__len__()):
    img, _ = dataset_test[elem]
    with torch.no_grad():
      prediction = model([img.to(device)])
      
    pred = prediction[0]['masks'][0, 0].mul(255).byte().cpu().numpy()
    true = dataset_test[elem][1]['masks'][0].byte().cpu().numpy()
    list_of_pred_mask.append(pred)
    list_of_true_mask.append(true)
    
  print(get_dice(list_of_true_mask,list_of_pred_mask))
  
```

```python
def get_list_prediction(dataset):
  """
  Get list prediction from test/valid dataset
     
    Parameters
    ----------
    dataset: (torch.utils.data.Dataset class) valid/test dataset
    
    Return:
    ----------
    list  which  is filled with arrays(np.array) 
    of prediction mask
  """
  model.eval()
  list_of_pred_mask = []
  
  for elem in range(len(dataset)):
    img, _ = dataset[elem]
    with torch.no_grad():
      prediction = model([img.to(device)])
    pred = prediction[0]['masks'][0, 0].mul(255).byte().cpu().numpy()
    list_of_pred_mask.append(pred)
    
  return list_of_pred_mask
```

**Посчитаем Dice для различных порогов**

```python
def dice(true, pred):
    EPS = 1e-10
    true = true.astype(bool)
    pred = pred.astype(bool)

    intersection = (true & pred).sum()
    im_sum = true.sum() + pred.sum()

    return 2.0 * intersection / (im_sum + EPS)
```
